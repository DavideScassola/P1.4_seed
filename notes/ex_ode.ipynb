```python
%matplotlib inline
from numpy import *
from matplotlib.pyplot import *
import scipy.linalg
import numpy.linalg

t0 = 0.
tf = 1.
y0 = 1.

s = linspace(t0,tf,5000)

exact = lambda x: x + exp(-x)
```

```python
h = 0.2
```

```python
def fe(y0,t0,tf,h):
    timesteps = arange(t0,tf+1e-10, h)
    sol = zeros_like(timesteps)
    sol[0] = y0
    for i in range(1,len(sol)):
        sol[i] = h*timesteps[i-1] + h + (1-h)*sol[i-1]
    
    return sol, timesteps

y, t = fe(l,y0,t0,tf,h)

_ = plot(t,y, 'o-')
_ = plot(s,exact(s))

error = numpy.linalg.norm(exact(t) - y, 2)
print(error)
```

```python
def be(y0,t0,tf,h):
    timesteps = arange(t0,tf+1e-10, h)
    sol = zeros_like(timesteps)
    sol[0] = y0
    for i in range(1,len(sol)):
        sol[i] = (h*timesteps[i] + h + sol[i-1])/(1+h)
    
    return sol, timesteps

y, t = be(l,y0,t0,tf,h)

_ = plot(t,y, 'o-')
_ = plot(s,exact(s))

error = numpy.linalg.norm(exact(t) - y, 2)
print(error)
```
