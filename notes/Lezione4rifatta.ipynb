```python
def interpolate(f,x,nodes):
    q = linspace(min(x),max(x),nodes)
    B = q.reshape((-1,1))**arange(nodes)
    Xvec = x.reshape(-1,1)**arange(nodes)
    return Xvec.dot(inv(B).dot(f(q)))
```

```python
def interpolate2(f,x,points):
    B = points.reshape((-1,1))**arange(len(points))
    Xvec = x.reshape(-1,1)**arange(len(points))
    return Xvec.dot(inv(B).dot(f(points)))
```

```python
def lagpol(x,nodes):
    q = linspace(min(x),max(x),nodes)
    B = q.reshape((-1,1))**arange(nodes)
    Xvec = x.reshape((-1,1))**arange(nodes)
    return Xvec.dot(inv(B))
```

```python
plot(x,lagpol(x,5))
```
